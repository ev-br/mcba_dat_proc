```python
import numpy as np

import numpy as np
import matplotlib
matplotlib.use('nbagg')

import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
from mcba.models.impurity import Par, SingleImpurity, energy
import mcba.db as db
```

```python
db_fname = '../data/g0.3/N15g0.05mq11.sqlite'
handle, = db.get_handles(db_fname)
par = db.get_param(handle)

enes = np.fromiter((energy(row, par) for row in db.row_iterator(handle)), dtype=np.float_ )
freqs = np.subtract.outer(enes, enes).ravel()

plt.hist(freqs, bins=20)
plt.show()
```
