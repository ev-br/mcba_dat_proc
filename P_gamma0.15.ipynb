```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')

import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
import mcba.db as db
from mcba.models.impurity import gamma, initial_q
import postproc as pp
```

```python
def read_data(base_path):
    r"""Read in the DB files from the path specified.
    
    Only read in the $t\to\infty$ momentum.
    """
    filez = [f for f in pp.list_dbs(base_path, lambda x: x.endswith('sqlite'))]

    data = pp.filter_pred(filez, lambda par, arr: True)

    return data
```

```python
def plot_momentum(data, ax=None):
    r""" Given 'data', plot $p_\infty$ vs $p_0$.
    """
    if ax is None:
        fig, ax = plt.subplots()

    for N in np.unique(data['N']):
        d = data[data['N'] == N]
        d = np.sort(d, order='q')

        ax.plot(d['q'], d['P1'], 'o-', lw=3)
        ax.plot(d['q'], d['P2'], 'o-',  label=str(N), lw=3)

    ax.set_xlabel(r'$q / k_F$')
    ax.set_ylabel(r'$P_\infty / k_F$')

    ax.set_title(r'$\gamma= %s $' % data['gamma'][0])
    ax.legend(loc='upper left')

    return fig, ax
```

```python
base_path = "../data/g0.3"
data = read_data(base_path)
fig, ax = plot_momentum(data)
plt.show()
```
